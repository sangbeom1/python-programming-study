#### 4.1 if문

- 비교 연산자 (<, <=, >, >=, ==, !=)

```python
1 < 2, 1 <= 2
```

```python
1 > 2, 1 >= 2
```

```python
1 == 2, 1 != 2
```

- 논리 연산자 (and, or, not)
- 참(True)은 숫자 1, 거짓(False)은 숫자 0에 대응됨
- 특별값 None은 거짓에 대응됨

```python
1 and 0, 1 & 0 # 둘 다 참이어야 참임
```

```python
(1 < 2) and (1 > 2)
```

```python
1 or 0, 1 | 0 # 둘 중 하나만 참이면 참임
```

```python
(1 < 2) or (1 > 2)
```

```python
not 1, not 0 # 거짓이면 참임
```

```python
not (1 < 2)
```

```python
not None
```

- 특별한 논리 연산자
- a **in** b: b 안에 a가 있으면 참
- a **not in** b: b 안에 a가 없으면 참

```python
'a' in 'ab', 'a' not in 'ab' # 문자열이 올 수 있음
```

```python
1 in [0,1], 'a' in ['a','b'], 1 not in [0,1], 'a' not in ['a','b'] # 리스트가 올 수 있음
```

```python
1 in (0,1), 'a' in ('a','b'), 1 not in (0,1), 'a' not in ('a','b') # 튜플이 올 수 있음
```

- if, else 조건문
- 각 제어문 뒤에는 콜론(:)이 들어감
- 각 제어문에 속하는 모든 문장에는 들여쓰기(indentation)를 해줘야 함

```python
score = 75
if score >= 90:
    print('A')
else:
    print('B')
```

```python
score = 75
if score >= 90:
    print('A')
else:
    if score >= 80:
        print('B')
    else:
        if score >= 70:
            print('C')
        else:
            print('D')
```

- if, elif, else 조건문
- 조건식이 여러 개 있을 때, 비효율적인 if 조건문을 elif 조건문으로 단순화  

```python
score = 75
if score >= 90:
    print('A')
elif score >= 80:
    print('B')
elif score >= 70:
    print('C')
else:
    print('D')
```

#### 4.2 try문

- try, except, else 조건문
- try 블록 수행 중 오류가 발생하면 except 블록을 수행
- try 블록 수행 중 오류가 발생하지 않으면 else 블록을 수행

```python
1/0 # 오류가 발생
```

```python
try:
    1/0
except:
    print('분모가 0이 나옴')
else:
    print('이상없음')
```

```python
try:
    0/1
except:
    print('분모가 0이 나옴')
else:
    print('이상없음')
```
