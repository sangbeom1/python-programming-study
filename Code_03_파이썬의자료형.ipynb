#### 3.5 딕셔너리(dictionary)

- 환율을 나타낼 때, "통화:기준율"의 관계로 "USD:1100", "JPY:900"과 같이 나타낼 수 있음
- 이러한 대응 관계를 나타낼 수 있는 자료형을 딕셔너리라고 함
- "통화"에 해당하는 것을 key, "기준율"에 해당하는 것을 value라고 함.
- 딕셔너리는 key와 그 key에 대응하는 value로 이루어져있고, key를 통해 value를 얻음
- 딕셔너리는 순서가 없는 자료형으로 key의 abc순으로 자동 정렬됨
- 딕셔너리에서 key는 고유한 값으므로 중복되는 key를 사용하면 안됨
- 딕셔너리의 key에는 수정이 가능한 리스트는 올 수 없으나, 수정이 불가능한 튜플은 올 수 있음

```python
a = {'USD':1100, 'JPY':900}
a['USD']
```

- 딕셔너리에 요소 추가하기

- 딕셔너리의 요소 삭제하기

#### 3.6 집합(set)

- 집합은 수학의 집합처럼 사용되는 자료형
- 딕셔너리 key와 마찬가지로 순서가 없으며, 중복을 허용하지 않음

```python
a = 'statistics'
s = set(a)
s
```

```python
a = ['사원', '대리', '과장', '차장', '부장']
s = set(a)
s
```

- 순서가 없는 자료형이므로 인덱싱을 위해서는 리스트나 튜플로 변환해야 함

```python
b = list(s)
b
```

```python
b[4]
```

- 교집합, 합집합, 차집합 구하기

```python
set({1,2,3})
```

```python
s1 = set([1,2,3])
s2 = set([2,4,6])
```

```python
s1 & s2 # 교집합
```

```python
s1 | s2 # 합집합
```

```python
s1 - s2 # 차집합
```

- 집합에 원소를 추가하기
- 중복된 원소는 허용하지 않음

```python
s = set([1,2,3])
s.update([3,4,5])
s
```

- 집합의 원소를 제거하기

```python
s = set([1,2,3])
s.remove(1)
s
```

#### 3.8 숫자형과 문자열의 출력

- 숫자형과 문자열 출력을 위해서 print 사용하고, 서식문자를 사용해 포맷팅 가능
- %d: 정수
- %f: 실수
- %s: 문자

```python
a = 100
b = 10
print('고객님께서 할인받으신 금액은 %d원입니다.' %a)
print('고객님께서 할인받으신 금액은 %d원이고, 사용하신 적립금은 %d원입니다.' %(a, b))
```

- %f는 default로 소수점 이하 6자리까지 나타냄

```python
a = 12.34
print('고객님께서 받으신 할인율은 %f%%입니다.' %a)
```

- %(총 자릿수).(소수점 이하 자릿수)f

```python
print('할인율 = %f' %a)
print('할인율 = %.3f' %a)
print('할인율 = %10.3f' %a)  # 뒤로 정렬 (default)
print('할인율 = %-10.3f' %a) # 앞으로 정렬
```

- %s는 숫자형을 문자로 취급해 그대로 출력
- 퍼센트(%) 기호는 서식문자와 구분하기 위해 기호 앞에 %를 한번 더 붙여줌

```python
print('할인율 = %s%%' %a)
```

```python
a = "박태영"
print('%s 고객님, 저희 매장을 방문해주셔서 감사합니다.' %a)
```
