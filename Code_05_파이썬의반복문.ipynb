#### 5.1 while문

- 조건문을 만족하는 동안 반복해서 블록을 수행
- a += b: a+b를 계산한 뒤, a+b의 값을 a에 할당
- a -= b: a-b를 계산한 뒤, a-b의 값을 a에 할당

```python
i = 0
while i < 10:
    i += 1
    print(i)
```

#### 5.2 for문

- 리스트 등의 첫 번째 요소부터 마지막 요소까지 차례로 대입하여 반복

```python
for i in [1,3,5,7,9]:
    print(i)
```

```python
dic = {'USD':1100, 'JPY':900, 'CNY':150, 'EUR':1300}
for key in dic.keys():
    print(key,dic[key])
```

- range(n): 0부터 (n-1)까지의 정수

```python
for i in range(10):
    print(i)
```

```python
j = 0
for i in range(10):
    j += i
print(j)
```

```python
for i in range(10):
    if i<=4:
        print(i)
    elif i>=6:
        print('over 5')
    else:
        print('equal to 5')
```

- range(a,b): a부터 (b-1)까지의 정수

```python
for i in range(5,10):
    print(i, end=" ")
```

- range(a,b,c): a에서부터 c의 간격으로 b보다 작은 정수

```python
for i in range(5,20,3):
    print(i, end=" ")
```

#### 4.3 중간에 멈추기와 건너뛰기

- break문
- for문이나 while문에서 반복을 하다가 중간에 반복을 멈추고 빠져나감

```python
for i in range(10):
    if i > 5:
        break
    print(i)
```

- continue문
- for문이나 while문에서 반복을 하다가 중간에 건너뜀

```python
for i in range(10):
    if (i%2) == 0:
        continue
    print(i)
```
