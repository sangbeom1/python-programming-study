#### 11.1 Matplotlib 모듈

- 그림을 그리기 위한 라이브러리
- 막대 그래프, 선 그래프, 히스토그램, 산점도 등

```python
from matplotlib import pyplot as plt # import matplotlib.pyplot as plt
```

- 막대그래프

```python
x = ['Yes','No']
y = [202,754]
z = [50,150]

plt.bar(x,y,yerr=z)

plt.title('Bar plot')
plt.xlabel('Default')
plt.ylabel('Count')

plt.show()
```

- 선 그래프 또는 시계열 그림

```python
x = ['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
y = [5.1,3.5,2.8,1.7,3.8,3,3.5,2.6,1.8,2.8,0.9]

plt.plot(x,y,'r--',x,y,'go')

plt.ylim([0,5.5])

plt.xlabel('Year')
plt.ylabel('Wages annual growth rate for civil servants')

plt.show()
```

- 히스토그램

```python
import numpy as np

x = np.random.randn(1000)

plt.hist(x, 10, density=True)

plt.xlabel('Normal')
plt.ylabel('Probability')

plt.axis([-4, 4, 0, 0.5])

plt.show()
```

- 산점도

```python
x = np.random.normal(size=1000)
y = 0.9*x + np.random.normal(size=1000)

plt.plot(x,y,'o')

plt.show()
```

- 함수의 그림

```python
def f(x):
    return np.cos(2*np.pi*x)**10

x = np.arange(0.0, 1.0, 0.01)

plt.plot(x, f(x), 'r--')

plt.show()
```

#### 11.2 Seaborn 모듈

- 고급 데이터 시각화를 위한 라이브러리
- https://seaborn.pydata.org/

```python
import seaborn as sns
```

```python
conda update seaborn --yes # seaborn을 업데이트 함 (아래 코드에서 경고/에러가 발생하는 경우 실행)
```

```python
x = np.random.normal(size=1000) # 표준정규분포에서 크기가 1000인 샘플 array 생성
y = 0.9*x + np.random.normal(size=1000) # x에 0.9를 곱하고, 오차 추가

sns.jointplot(x=x,y=y, color="#345beb") # 로얄블루 색상

plt.show()
```

```python
sns.jointplot(x=x,y=y, kind="hex", color="#345beb")

plt.show()
```
